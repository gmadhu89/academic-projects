Data cleaning

```python
import pandas as pd
import numpy as np
```

```python
playtime_df = pd.read_csv('D:/PhD VIII/CSE 6242/Proj_6242/CSE-6242-Project/DB/playtime_by_country.csv')
game_df = pd.read_csv('D:/PhD VIII/CSE 6242/Proj_6242/CSE-6242-Project/DB/steam_vg_merged.csv')
```

```python
merged_df = playtime_df.merge(game_df,
                 how='left',
                 left_on='game',
                 right_on='Name')
```

```python
merged_df.to_csv('D:/PhD VIII/CSE 6242/Proj_6242/CSE-6242-Project/DB/playtime_all.csv')
```

```python
col_names=['game', 'Genre', 'tags', 'ESRB_Rating', 'Platform', 'price',
           'Critic_Score',
           'metacritic', 'recommendations', 'positive', 'negative',
           'country_code', 'time']
sheared_df = merged_df[col_names]
```

```python
sheared_df.to_csv('D:/PhD VIII/CSE 6242/Proj_6242/CSE-6242-Project/DB/playtime_cleaned.csv', index=False)
```
